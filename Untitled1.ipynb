```python
import pandas as pd
```

```python
df = pd.read_csv('students.csv')
```

```python
df.head()
df.shape
df.isnull().sum()
```

```python
df.duplicated()
df.drop_duplicates()
df.dtypes
```

```python
from numpy import astype
df['Total marks'] = df['Math'] + df['Physics'] + df['Chemistry']
df['Average'] = (df['Total marks']/3).round().astype(int)
```

```python
def grade_calc(row):
    avg = row['Average']

    if avg >= 85:
        return "A"
    elif avg >= 70:
        return "B"
    elif avg >= 50:
        return "C"
    else:
        return "D"
```

```python
df['Grade'] = df.apply(grade_calc, axis=1)
df.head(10)
```

```python
import matplotlib.pyplot as plt
plt.bar(df['Name'], df['Average'])
```

```python
# calculate average of each subject
subject_averages = df[['Math','Physics','Chemistry']].mean()

# plot bar chart
plt.figure() #makees sure ur plot is on a new canvas
plt.bar(subject_averages.index, subject_averages.values)


# labels
plt.title("Average Marks per Subject")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")

plt.show()
```

```python
def result_calc(row):
  if row['Grade'] in ['C', 'D']:
    return 'Fail'
  else:
    return 'Pass'
df['Actual_Result'] = df.apply(result_calc , axis = 1)
df.head(15)
```

```python
result= df['Actual_Result'].value_counts()
print(result)
```

```python
plt.figure()
plt.bar(result.index, result.values)
plt.title('Result Analysis')
plt.xlabel('Result')
plt.ylabel('Number of Students')
plt.show()
```

```python
df['Predicted_Result'] = (
    (df['Average'] >= 50) &
    (df['Attendance'] >= 75)
).map({True:'Pass', False:'Fail'})
```
